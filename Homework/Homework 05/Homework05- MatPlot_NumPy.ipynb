````python
# Histogram
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.randn(1000)  # 1000 random values from a normal distribution

# Create the histogram
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='blue', alpha=0.7, edgecolor='black')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram Example")

# Show the plot
plt.show()
````

````python
# Line Plot
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import numpy as np  # Import NumPy for numerical operations

# Generate sample data
x = np.linspace(0, 10, 100)  # Create 100 evenly spaced values from 0 to 10
y = np.sin(x)  # Compute the sine of each x value

# Create the line plot
plt.figure(figsize=(8, 5))  # Set the figure size (width=8, height=5)
plt.plot(x, y, color='blue', linewidth=2, linestyle='-', label='sin(x)')  
# - Plots x vs y
# - Sets line color to blue
# - Line width is 2
# - Line style is solid '-'
# - Adds a label for the legend

# Add labels and title
plt.xlabel("X-axis")  # Label for the x-axis
plt.ylabel("Y-axis")  # Label for the y-axis
plt.title("Line Plot Example")  # Title of the plot

# Add a legend (to display the label set in plt.plot)
plt.legend()

# Show the plot
plt.show()
````

````python
# Scatter Plot
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import numpy as np  # Import NumPy for numerical operations

# Generate random data for the scatter plot
x = np.random.rand(100) * 10  # Generate 100 random x values between 0 and 10
y = np.random.rand(100) * 10  # Generate 100 random y values between 0 and 10
colors = np.random.rand(100)  # Assign random colors to each point
sizes = np.random.rand(100) * 200  # Assign random sizes to each point

# Create the scatter plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.scatter(x, y, c=colors, s=sizes, alpha=0.7, cmap='viridis', edgecolors='black')
# - x, y: Coordinates of the points
# - c=colors: Color each point randomly
# - s=sizes: Vary point size
# - alpha=0.7: Set transparency to 70%
# - cmap='viridis': Use the 'viridis' colormap
# - edgecolors='black': Set black borders for visibility

# Add labels and title
plt.xlabel("X-axis")  # Label for the x-axis
plt.ylabel("Y-axis")  # Label for the y-axis
plt.title("Scatter Plot Example")  # Title of the plot

# Show the plot
plt.show()
````

````python
# Pie Plot
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

# Define data for the pie chart
labels = ['Apple', 'Banana', 'Cherry', 'Date']  # Categories
sizes = [30, 25, 35, 10]  # Percentage distribution of each category
colors = ['red', 'yellow', 'purple', 'brown']  # Colors for each segment
explode = (0, 0.1, 0, 0)  # "Explode" the second slice for emphasis

# Create the pie chart
plt.figure(figsize=(7, 7))  # Set figure size
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True)
# - sizes: Determines the size of each slice
# - labels: Assigns names to each slice
# - colors: Sets different colors for each slice
# - autopct='%1.1f%%': Displays the percentage value on each slice
# - startangle=140: Rotates the pie chart to start from a specific angle
# - explode: Highlights the second slice (Banana)
# - shadow=True: Adds a shadow effect

# Add a title
plt.title("Fruit Distribution Pie Chart")

# Show the plot
plt.show()
````

````python
# Area Plot
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import numpy as np  # Import NumPy for numerical operations

# Generate sample data
x = np.linspace(0, 10, 100)  # Create 100 evenly spaced values from 0 to 10
y1 = np.sin(x)  # Compute sine function
y2 = np.cos(x)  # Compute cosine function

# Create the area plot
plt.figure(figsize=(8, 5))  # Set figure size
plt.fill_between(x, y1, color='skyblue', alpha=0.6, label='sin(x)')  
# - Fills the area under the sin(x) curve
# - Color is set to sky blue
# - Alpha controls transparency (0.6 = 60% opaque)

plt.fill_between(x, y2, color='orange', alpha=0.6, label='cos(x)')  
# - Fills the area under the cos(x) curve
# - Color is set to orange

# Add labels and title
plt.xlabel("X-axis")  # Label for the x-axis
plt.ylabel("Y-axis")  # Label for the y-axis
plt.title("Area Plot Example")  # Title of the plot

# Add a legend
plt.legend()

# Show the plot
plt.show()
````

````python
# 3d Plot - Bar Plot
import matplotlib.pyplot as plt  # Import Matplotlib for plotting
import numpy as np  # Import NumPy for numerical operations
from mpl_toolkits.mplot3d import Axes3D  # Import 3D plotting toolkit

# Generate sample data
x = np.arange(4)  # X-axis positions (categories)
y = np.zeros(4)  # Y-axis positions (set to zero for a simple bar chart)
z = np.zeros(4)  # Z-axis positions (base of bars)

dx = np.ones(4)  # Width of bars
dy = np.ones(4)  # Depth of bars
dz = [10, 20, 15, 30]  # Heights of the bars

categories = ['A', 'B', 'C', 'D']  # Labels for the categories

# Create the 3D figure
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')  # Add a 3D subplot

# Create the 3D bar plot
ax.bar3d(x, y, z, dx, dy, dz, color=['red', 'blue', 'green', 'purple'])
# - x, y, z: Positions of bars
# - dx, dy: Width and depth of bars
# - dz: Height of bars
# - color: Different colors for each bar

# Set axis labels
ax.set_xlabel("Category")  # Label for X-axis
ax.set_ylabel("Y")  # Label for Y-axis (not very relevant here)
ax.set_zlabel("Value")  # Label for Z-axis (bar heights)

# Set X-axis category labels
ax.set_xticks(x)
ax.set_xticklabels(categories)

# Set title
plt.title("3D Bar Plot Example")

# Show the plot
plt.show()
````

````python
import numpy as np
import csv

# Define correct data types
dtype = [
    ('Name', 'U100'),
    ('Year_of_Release', 'i4'),
    ('Genre', 'U20'),
    ('Publisher', 'U50'),
    ('NA_Sales', 'f4'),
    ('EU_Sales', 'f4'),
    ('JP_Sales', 'f4'),
    ('Other_Sales', 'f4'),
    ('Global_Sales', 'f4'),
    ('Critic_Score', 'f4'),
    ('Critic_Count', 'i4')
]

# Read CSV file correctly, handling quoted strings
cleaned_data = []
with open('vgsales.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    headers = next(reader)  # Skip the header row
    for row in reader:
        if len(row) < 11:  # Skip incomplete rows
            continue
        
        # Extract and clean data
        name = row[0].strip('"')  # Remove extra quotes around names
        try:
            year = int(row[1]) if row[1].isdigit() else 0  # Handle missing years
        except ValueError:
            year = 0  # Set to 0 if not an integer
        
        # Convert numeric fields safely
        try:
            na_sales = float(row[4]) if row[4] else 0
            eu_sales = float(row[5]) if row[5] else 0
            jp_sales = float(row[6]) if row[6] else 0
            other_sales = float(row[7]) if row[7] else 0
            global_sales = float(row[8]) if row[8] else 0
            critic_score = float(row[9]) if row[9] else 0
            critic_count = int(row[10]) if row[10].isdigit() else 0
        except ValueError:
            continue  # Skip rows with invalid numbers
        
        # Append cleaned data
        cleaned_data.append((name, year, row[2], row[3], na_sales, eu_sales, jp_sales, other_sales, global_sales, critic_score, critic_count))

# Convert to NumPy structured array
data = np.array(cleaned_data, dtype=dtype)

# Sort by Name, Year, and NA_Sales
sorted_data = np.sort(data, order=['Name', 'Year_of_Release', 'NA_Sales'])

# Display first 10 rows for verification
print(sorted_data[:10])
````
